```python
import matplotlib.pyplot as plt

# gNB numbers
gnb_numbers = list(range(1, 21))

# Connection times in ms
connection_times = [
    0.35, 0.36, 0.37, 0.44, 0.24, 0.40, 0.20,
    0.20, 0.15, 0.21, 0.08, 0.25, 0.07, 0.21,
    0.12, 0.13, 0.09, 0.19, 0.19, 0.18
]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(gnb_numbers, connection_times, marker='o', linestyle='-', color='blue')
plt.title('gNB Connection Latency')
plt.xlabel('gNB Number')
plt.ylabel('Connection Time (ms)')
plt.grid(True)
plt.xticks(gnb_numbers)
plt.show()
```
