```python
import matplotlib.pyplot as plt

# GNB numbers (incremented each time)
gnb_numbers = list(range(1, 21))

# Corresponding connection times in ms
connection_times = [
    0.14, 0.80, 0.20, 0.34, 0.39, 0.31, 0.38,
    0.30, 0.12, 0.19, 0.16, 0.14, 0.21, 0.13,
    0.37, 0.44, 0.16, 0.15, 0.34, 0.25
]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(gnb_numbers, connection_times, marker='o', linestyle='-', color='green')
plt.title('gNB Connection Latency Over Time')
plt.xlabel('gNB Connection Number')
plt.ylabel('Connection Time (ms)')
plt.grid(True)
plt.xticks(gnb_numbers)
plt.show()
```
